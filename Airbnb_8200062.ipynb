## Availability per day

For this task, we want to create a plot with the number of listings available per day, on the  𝑦 axis, and the date, on the  𝑥 axis.

* First, let's read our data by loading the necessary csv file that contains the availability of each listing for each day.

```python
av = pd.read_csv('detailed_calendar.csv')
av
```

* We are going to use the to_datetime method to convert the date column to a datetime object and then sort the df by the date column. 

```python
av['date'] = pd.to_datetime(av['date'])
av = av.sort_values('date')
```

* Now that the df is sorted by the date (meaning that for the first date of the df, all the listings appear, then for the second date and so on) we want to take only the available listings ( available = true) and group them by date. We also use the size method to to get the frequency of the available listings for each date. Finally, with the reset_index method we set the index name of the column as count so that we can refer to it later.

```python
available_listings = av[av['available'] == 't']
daily = available_listings.groupby('date').size().reset_index(name='count')
daily
```

To create the plot we are going to import the `matplotlib.pyplot` library which is a very popular visualisation library in python. 
- Figsize refers to the width and height of the plot in inches. 
- In the plot method, we take the two columns that we need from the df, date and count, to create the plot.
- With title, xlabel and ylabel we name the plot and the label of the axes.
- The xticks is responsible for the labels in the x axis. The daily['date'] states that we select elements from this column and the [::int(len(daily['date'])/10)] divides the total length of the 'date' column by 10 and converts the result to an integer.



```python
import matplotlib.pyplot as plt
import numpy as np

#Convert the 'date' and 'count' columns to a numpy array
date_array = np.array(daily['date'])
count_array = np.array(daily['count'])

# Plotting the data
plt.figure(figsize=(12, 6))
plt.plot(date_array, count_array, marker='o', color='b')
plt.title('Availability per day')
plt.xlabel('Date')
plt.ylabel('Available listings')
plt.xticks(daily['date'][::int(len(daily['date'])/10)],rotation=45)
plt.grid(True)


# Display the plot
plt.show()
```

If we take a glance at the results of the plot, we can see that they make sense. 
- First of all, until November the availability is significantly lower compared to the other months, which makes sense as people are more likely to book a listing for the following months rather that one year from now. 
- We can also see that as the weather gets warmer (after March) the availability gradually decreases as tourists start visiting the country. 
- Finally, it is worth noticing that there is a sudden drop in availability between May and July, mainly because in June, Athens has very warm weather without heatwaves, making the city a very popular destination for tourists that month!

## Reviews, Occupacy per Listing, Average Income

We need to create a histogram, where on the x axis will be the number of reviews binned from the lowest number to the highest rounded up to the nearest hundred, and on the y axis the frequency of each bin - meaning how listings each bin contains.
As the listing id and the number of reviews are included in the listings_csv, we will not load a different csv, we will only create a new dataframe containing these two columns.

```python
reviews = pd.DataFrame({'id': df['id'], 'no_of_reviews': df['number_of_reviews']})
reviews
```

1. Now that we have all listings id and the number of reviews per listing, we can proceed with the <b>binning of the reviews</b> and the <b>creation of the histogram</b> by using the matplotlib library.
* First, we need to calculate the bins. For this, we are going to select the number of reviews column and get the minimum and maximum values.
* To round up the maximum value to the closest hundred, we need to divide the maximum value by 100 and get the modulus of the division. Then we will add 1 and multiply by 100 to get the next 100 of the number that we had.
* From the numpy library, we will use the linspace methods, which takes a start and a stop value, as well as a n step value, and returns n equal bins. For this assignment, we will use 20 bins, however, we can use as many we want, depending on how we want the histogram to look like.
* Finally, we use the same matplotlib commands to make a histogram, except that we now load the bins variable so that the histogram separates the data into the bins that we made before.

```python
# Calculating the range for binning
min_reviews = reviews['no_of_reviews'].min()
max_reviews = reviews['no_of_reviews'].max()
max_rounded = (max_reviews // 100 + 1) * 100
import numpy as np
bins = np.linspace(min_reviews,max_rounded,num=20)

# Creating the histogram
plt.figure(figsize=(8, 6))
plt.hist(reviews['no_of_reviews'], bins=bins, edgecolor='black')
plt.title('Number of Reviews per Listing')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

From the histogram that we made we can see that the majority of the listings belong to the first bin (more than 80% of the listings) where the number of reviews ranges from zero which is the minimum value to approximately 50 reviews. As a result, the first histogram is so large that we cannot clearly the other ones as they are way smaller compared to the first one. This is the reason why we are going to make a subset of the histogram, only this time we will exclude all listings that have less than 50 reviews so that we can see more clearly the listings will lots of reviews (the other 20% approx. of the data).

```python
# Calculating the range for binning
max_reviews = reviews['no_of_reviews'].max()
max_rounded = (max_reviews // 100 + 1) * 100
import numpy as np
bins = np.linspace(50,max_rounded,num=20)

# Creating the histogram
plt.figure(figsize=(8, 6))
plt.hist(reviews['no_of_reviews'], bins=bins, edgecolor='black')
plt.title('Number of Reviews per Listing')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Now we can see better what is happening with the listings with more than 50 reviews each! The majority of these ranges from 50-200 reviews,however, we can see that there are some listings-very few- that have over 600 reviews each one of them.

2. Moving on, we are going to <b>calculate the average occupacy per listing per month</b>. First we are going to read a csv file containing the date of each review for all listings.

```python
f = pd.read_csv('reviews.csv')
f
```

Using the todatetime method we are going to convert the column to a date form so that then we can extract only the month from the date column and create a new month column.

```python
f['date'] = pd.to_datetime(f['date'])

# Extract the month from the 'date' column
f['month'] = f['date'].dt.month

f
```

We are now going to group the df by month and apply the size method, so that we get the total reviews per month for all listings.

```python
fnew = f.groupby('month').size()
fnew
```

Now, let's convert the series to a DataFrame and name the columns month and reviews so that the data can be understood easily.

```python
fnew = fnew.reset_index()
fnew.columns = ['month', 'reviews']
fnew
```

We will now find how many the unique listings are. <br>
We can use the `f.drop_duplicates(subset='listing_id').shape[0]` to get the number of all unique listings but the reviews.csv file contains *only the listings that have at least 1 review*, leading to misleading results. <br>That's why we will use the listings csv file which contains all the listings id, whether they have reviews or not. The listings file doesn't contain any duplicate id values, however we can apply the drop_duplicates method here just to be sure.

```python
#checking for duplicate id values
unique_listings = df.drop_duplicates('id').shape[0]
unique_listings
```

Now, we can divide the reviews per month with the number of listings, to find the average reviews per month per listing. We also append the new column with the average reviews per listing for each month to the dataframe.

```python
av_reviews = fnew['reviews']/unique_listings
fnew['av_reviews_per_listing'] = av_reviews
fnew
```

Let's assume that half the bookings result to a review. So, if we multiply the average reviews per listing by 2 we can get the average bookings per listing per month, and then append it to the dataframe as a new column.

```python
bookings = 2 * fnew['av_reviews_per_listing']
fnew['av_bookings_per_listing'] = bookings
fnew
```

We have now found the average occupacy per listing per month and we can see that the results make sense as during the summer months the occupacy in Athens is much higher as Athens is a summer destination and that's when tourists flee to the capital of Greece. The maximum occupacy is 12.8 days (almost half of the month) during August and the lowest is in December where the occupacy is just 4 days. <br><br> 
If we take a closer look, we can see that the data from this df complements the data from the "Availability per day" as we found that the availability during summer is lower, which agrees with the occupacy in this df where it is high during the summer months.

Now, from the listings csv file, we will choose the price column and find the average price per listing, by applying the `mean()` method. We are also assuming that the price is standard for each listing throughout the year. However, in real-life scenarios prices tend to change on a daily basis or according to demand.

```python
av_price_per_listing = df['price'].mean()
av_price_per_listing
```

Let's also assume that each booking refers to a 3 nights stay. That means that the `income per booking is 3 multiplied by the av_price_per_listing`. Then, we can find the average income per month if we multiply the income with the average_bookings per month of the df.

```python
average_income = 3 * av_price_per_listing * fnew['av_bookings_per_listing']
fnew['av_income_per_listing'] = average_income
fnew
```

<u>It's important to note that</u> we have made a hypothesis where each booking is for a 3 nights stay. However, this is not feasible in some cases. For example, in June-September where the average bookings are approx. 12 days per month, the output is 12 * 3 = 36 days per month which exceed the 30-31 day limit that each month has. Therefore, the average income in some cases is lower that what we have found.

## Prices per Day

We want to find the average price per listing per day and then create a plot with the data. First let's read the respective csv file.

```python
data = pd.read_csv('detailed_calendar.csv')
data
```

By using the `info()` method we can see that both columns we need - date and price - are not in the appropriate format, thus, we should first convert our data. 

```python
data.info()
```

We will convert the date column from object to date format by using the `to_datetime()` method. We will also remove the dollar sign by using the `replace()` method from the price column and convert the data to numerical format by applying the `to_numeric()` method which converts an argument to numeric type. <br>

The `replace()` method is applied on the price column of the dataframe and it searches for any dollar signs or commas and replaces them with nothing, meaning that it removes them from the column. As this is a regular expression "[\ $,]" , we have also added the regex = True argument, indicating that the function should interpret the search pattern as a regular expression.
<br>

After we edited the data, we will *group them by the date and apply the mean() method for the price column* so that we get the average price per listing per day.

```python
data['date'] = pd.to_datetime(data['date'])
data['price'] = pd.to_numeric(data['price'].replace('[\$,]', '', regex = True))
mean_price_per_date = data.groupby('date')['price'].mean()
mean_price_per_date = pd.DataFrame(mean_price_per_date)

#resetting the index and renaming the columns so that we refer to them later
mean_price_per_date.reset_index(inplace=True)
mean_price_per_date.columns = ['date', 'price']
mean_price_per_date
```

Now that we have created a table with the average price per listing per day, let's create a plot to visualise the data.

```python
#convert the columns to numpy arrays
date_array = np.array(mean_price_per_date['date'])
price_array = np.array(mean_price_per_date['price'])

plt.figure(figsize=(10, 6))
plt.plot(date_array, price_array, marker='o', linestyle='-')
plt.title('Average Price Per Listing')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

As we can see from the plot, the prices during the summer are much higher than the rest of the year, which completely makes sense with the above findings.

## Visualise Listings

For the final part of this assignment, we must draw an interactive map to show the listings ovelayed on Athens. 
* To implement this, we use the folium library and we load a map with the coordinates that belong to Athens. 
* After that, we create a loop for every row of the dataframe of the listings file, and we will read the latitude and longitude of each listing. 
* In this way, we can add a marker on the Athens map for every listing of the dataframe and with the popup argument we can choose something to display on the marker when you click on it. Here we load the name column of the dataframe so in each marker some basic information of the listing appear to the user. 
* Finally, due to the fact that the name column contains a star symbol that might cannot be read by a browser, we will use the replace method so that any '★' characters are replaced with the HTML entity '☆'.
* We will change the default blue color to green by assigning the green value to a variable and then using the icon=folium.Icon(color=color) command.
*  We are also saving the map to an external html file that can be opened by the browser as the map is heavy on resources and most probably it would not load on jupyter notebook.

<br>*Please note that because of the heavy resources that the folium map needs, the html file might need some time to load or the page might need to be refreshed before it loads correctly.*

```python
import folium, html 

m = folium.Map(location=(37.9838, 23.7275), zoom_start = 13, max_zoom=15) #setting the maximum level that the user can zoom in

for i, row in df.iterrows():
    
    color = 'green'  

    folium.Marker([row['latitude'], row['longitude']], popup=row['name'].replace('★', '&star;'), icon=folium.Icon(color=color)).add_to(m)

m.save("athens_listings.html")
```
